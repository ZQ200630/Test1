```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x**2)
plt.plot(x, y, label="sin(x)")
plt.plot(x, z, "r--", label="cos(x^2)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("A Simple Plot")
plt.legend()
plt.show()
```

```python
# Lorentzian line shape
def lorentzian(x, x0, gamma):
    return gamma / np.pi / ((x - x0)**2 + gamma**2)

# Generate data
x = np.linspace(-5, 5, 1000)
y = lorentzian(x, 1, 1)

# Plot data
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Lorentzian")
plt.show()
```

```python
def test_lorentzian():
    x = np.linspace(-5, 5, 1000)
    y = lorentzian(x, 1, 1)
    assert np.all(y >= 0)
```
