```python
import numpy as np
from matplotlib import pyplot as plt
from celluloid import Camera
```

```python
A = 1 # 振幅
w1 = 20# 角频率
w2 = 22
n1 = 1.01# 折射率
n2 = 1.06
k1 = n1*10# 波数
k2 = n2*11

x = np.linspace(0, 6*np.pi, 1000)
t = np.linspace(0, 2*np.pi, 100)

fig,axes = plt.subplots(3,1,figsize=(10, 5))
camera = Camera(fig)

for ti in t:
    yi1 = A*np.sin(k1*x - w1*ti)
    yi2 = A*np.sin(k2*x - w2*ti)
    yi = yi1+yi2
 
    line1, = axes[1].plot(x, yi1, '#EF9A9A')
    line2, = axes[2].plot(x, yi2, '#A5D6A7')
    line_sum, = axes[0].plot(x, yi, '#64B5F6')
    axes[0].legend([line_sum],['y1+y2'],loc='upper right')
    axes[1].legend([line1],['y1'], loc='upper right')
    axes[2].legend([line2],['y2'], loc='upper right')
 
    camera.snap()
animation = camera.animate()
animation.save('with_dispersion.gif', writer='imagemagick')
```
